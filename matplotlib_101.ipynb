# Matplotlib 101

```python
from matplotlib import pyplot as plt
```

## Scatter Plot

```python
X = [1, 2, 3, 4, 5, 6]
Y = [60, 100, 120, 150, 210, 250]

plt.scatter(X,Y)
plt.show()
```

### Connect Scatter Plot with Lines

```python
# Connect Scatter Plot with Lines
plt.scatter(X,Y)
plt.plot(X,Y)
plt.show()
```

### Add Axis Titles

```python
plt.scatter(X,Y)
plt.plot(X,Y)
plt.title("Odometer vs. Time")
plt.xlabel("Time (h)")
plt.ylabel("Odometer (km)")
plt.show()
```
